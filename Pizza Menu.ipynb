```python
#Creating the Pizza Class

class Pizza:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price
```

```python
#Defining Subclasses of Pizza

class Margherita(Pizza):
    def __init__(self):
        super().__init__("Margherita", "Tomato sauce, mozzarella cheese, basil", 135.00)

class Supreme(Pizza):
    def __init__(self):
        super().__init__("Supreme", "Tomato sauce, sliced mozzarella, pepperoni, sausage, onions, peppers, mushrooms", 195.00)
        
class Marinara(Pizza):
    def __init__(self):
        super().__init__("Marinara", "Tomato sauce, garlic, oregano", 120.00)  
        
class Chicago(Pizza):
    def __init__(self):
        super().__init__("Chicago", "Tomato sauce, ground beef, sausage, pepperoni, mushrooms, green peppers, parmesan cheese", 215.00)         

class Sicilian(Pizza):
    def __init__(self):
        super().__init__("Sicilian", "Tomato, onion, anchovies, herbs, caciocavallo", 175.00) 
    
class Greek(Pizza):
    def __init__(self):
        super().__init__("Greek", "Pizza sauce, dried oregano, part-skim mozzarella, roasted red peppers, artichoke, kalamata olives, crumbled feta", 170.00)
        
class California(Pizza):
    def __init__(self):
        super().__init__("California", "Chicken, peanut sauce, artichoke hearts, goat cheese", 180.00)
        
class Diavola(Pizza):
    def __init__(self):
        super().__init__("Diavola", "Tomato sauce, mozzarella cheese, spicy salami, hot chili peppers", 190.00)
    
class Hawaiian(Pizza):
    def __init__(self):
        super().__init__("Hawaiian", "Tomato sauce, mozzarella cheese, beef bacon, and pineapple", 180.00)
        
class Veggie(Pizza):
    def __init__(self):
        super().__init__("Veggie", "Green bell pepper, red onion, mushrooms, tomato, olives, zucchini, eggplants", 160.00)
```

```python
#Defining Extra Ingredients

class Ingredient:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Olives(Ingredient):
    def __init__(self):
        super().__init__("Olives", 4.00)

class Mushrooms(Ingredient):
    def __init__(self):
        super().__init__("Mushrooms", 10.00)
        
class Parmesan(Ingredient):
    def __init__(self):
        super().__init__("Parmesan", 30.00)
        
class Bacon(Ingredient):
    def __init__(self):
        super().__init__("Bacon", 40.00)

class Pineapple(Ingredient):
    def __init__(self):
        super().__init__("Pineapple", 20.00)

class Zucchinis(Ingredient):
    def __init__(self):
        super().__init__("Zucchinis", 12.00)

class Eggplants(Ingredient):
    def __init__(self):
        super().__init__("Eggplants", 12.00)
        
class Cheddar(Ingredient):
    def __init__(self):
        super().__init__("Cheddar", 20.00)

class Pepperoni(Ingredient):
    def __init__(self):
        super().__init__("Pepperoni", 40.00)

class Sausage(Ingredient):
    def __init__(self):
        super().__init__("Sausage", 35.00)
        
class Peppers(Ingredient):
    def __init__(self):
        super().__init__("Peppers", 8.00)

class Chicken(Ingredient):
    def __init__(self):
        super().__init__("Chicken", 35.00)

class Salami(Ingredient):
    def __init__(self):
        super().__init__("Salami", 32.00)

class Onions(Ingredient):
    def __init__(self):
        super().__init__("Onions", 10.00)
```

```python
#Define a function to display the menu and extras to the customer. This function should prompt the customer to select a pizza, and then offer the option to add extra ingredients. 

def display_menu():
    print("Welcome to Pizza Python!")
    print("Here are the pizza options:")
    print("1. Margherita - Tomato sauce, mozzarella cheese, basil - ₺135.00")
    print("2. Supreme - Tomato sauce, mozzarella cheese, pepperoni, sausage, onions, peppers, mushrooms - ₺195.00")
    print("3. Marinara - Tomato sauce, garlic, oregano - ₺120.00")
    print("4. Chicago - Tomato sauce, ground beef, sausage, pepperoni, mushrooms, green peppers, parmesan cheese - ₺215.00")         
    print("5. Sicilian - Tomato, onion, anchovies, herbs, caciocavallo - ₺175.00")
    print("6. Greek - Pizza sauce, dried oregano, part-skim mozzarella, roasted red peppers, artichoke, kalamata olives, crumbled feta - ₺170.00")
    print("7. California - Chicken, peanut sauce, artichoke hearts, goat cheese - ₺180.00")
    print("8. Diavola - Tomato sauce, mozzarella cheese, spicy salami, and hot chili peppers - ₺190.00")
    print("9. Hawaiian - Tomato sauce, mozzarella cheese, beef bacon, and pineapple - ₺180.00")
    print("10. Veggie - Green bell pepper, red onion, mushrooms, tomato, olives, zucchini, eggplants - ₺160.00")

    pizza_choice = input("Please select a pizza by entering its number: ")

    if pizza_choice == "1":
        pizza = Margherita()
    elif pizza_choice == "2":
        pizza = Supreme()
    elif pizza_choice == "3":
        pizza = Marinara()
    elif pizza_choice == "4":
        pizza = Chicago()
    elif pizza_choice == "5":
        pizza = Sicilian()
    elif pizza_choice == "6":
        pizza = Greek()
    elif pizza_choice == "7":
        pizza = California()
    elif pizza_choice == "8":
        pizza = Diavola()
    elif pizza_choice == "9":
        pizza = Hawaiian()
    elif pizza_choice == "10":
        pizza = Veggie()
    else:
        print("Invalid selection!")
        return

    print("Great choice! Here are the extra ingredients you can add:")
    
    print("1. Olives - ₺4.00")
    print("2. Mushrooms - ₺10.00")
    print("3. Parmesan - ₺30.00")
    print("4. Bacon - ₺40.00")
    print("5. Pineapples - ₺20.00")
    print("6. Zucchinis - ₺12.00")
    print("7. Eggplants - ₺12.00")
    print("8. Cheddar - ₺20.00")
    print("9. Pepperoni - ₺40.00")
    print("10. Sausage - ₺35.00")
    print("11. Peppers - ₺8.00")
    print("12. Chicken - ₺35.00")
    print("13. Salami - ₺32.00")
    print("14. Onions - ₺10.00")

    extras_choice = input("Would you like to add any extras? (y/n)")

    if extras_choice.lower() == "y":
        extras = []
        while True:
            extra_choice = input("Please select an extra by entering its number, or enter 'done' when you're finished: ")
            if extra_choice == "1":
                extras.append(Olives())
            elif extra_choice == "2":
                extras.append(Mushrooms())
            elif extra_choice == "3":
                extras.append(Parmesan())
            elif extra_choice == "4":
                extras.append(Bacon())
            elif extra_choice == "5":
                extras.append(Pineapples())
            elif extra_choice == "6":
                extras.append(Zucchinis())
            elif extra_choice == "7":
                extras.append(Eggplants())
            elif extra_choice == "8":
                extras.append(Cheddar())
            elif extra_choice == "9":
                extras.append(Pepperoni())
            elif extra_choice == "10":
                extras.append(Sausage())
            elif extra_choice == "11":
                extras.append(Peppers())
            elif extra_choice == "12":
                extras.append(Chicken())
            elif extra_choice == "13":
                extras.append(Salami())
            elif extra_choice == "14":
                extras.append(Onions())    
            elif extra_choice.lower() == "done":
                break
            else:
                print("Invalid selection.")
    else:
        extras = []

    return pizza, extras
```

```python
#Define a function to calculate the total price of the customer's order, including any extras they have selected. 

def calculate_price(pizza, extras):
    total_price = pizza.price
    for extra in extras:
        total_price += extra.price
    return total_price
```

```python
#Define a function to collect the customer's information, including their ID, name, credit card number, and password. For simplicity, we'll just use the input function to get this information from the customer. However, in a real application, you would want to use a more secure method to collect and store this information. 

def collect_information():
    # Collect customer ID
    customer_id = input("Enter your customer ID: ")

    # Collect customer name
    customer_name = input("Enter your name: ")

    # Collect credit card information
    print("Enter your credit card information:")
    credit_card_number = input("Credit card number: ")
    credit_card_password = input("Credit card password: ")

    return customer_id, customer_name, credit_card_number, credit_card_password
```

```python
import csv
from datetime import datetime

def collect_information(pizza, extras, total_price):
    customer_id = input("Please enter your customer ID: ")
    customer_name = input("Please enter your name: ")
    credit_card_number = input("Please enter your credit card number: ")
    password = input("Please enter your password: ")
    order_description = f"{pizza.name} pizza with "
    for extra in extras:
        order_description += f"{extra.name}, "
    order_description = order_description.rstrip(", ")
    order_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    with open("Orders_Database.csv", "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([customer_id, customer_name, credit_card_number, password, order_description, total_price, order_time])

    print("Thank you for your order!")
```

```python
def main():
    
    pizza, extras = display_menu()
    total_price = calculate_price(pizza, extras)
    print(f"The total price of your order is ${total_price:.2f}.")
    collect_information(pizza, extras, total_price)
```
